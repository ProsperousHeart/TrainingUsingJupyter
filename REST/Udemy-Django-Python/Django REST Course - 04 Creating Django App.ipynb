## Option 1:  venv (built-in)

Inorder to set up your virtual environment, you will need to connect to the Vagrant server in your **git bash** by:

1. going to your project directory
2. running `vagrant ssh`
3. change directory into the vagrant directory:  `cd /vagrant/`
4. run:  `python -m venv PATH\TO\VIRTUAL\ENV` such as `python -m venv ~/env`

The directory in #4 is where all of the dependencies you install via the python package manager will be stored. In the above example, it will create a new folder in the Vagrant home directory. Doing it this way will set up the environemtn in the server and not sync it with your original machine. This way if you ever need to destroy & create the Vagrant server frm scratch you can do so with a fresh virtual environment.

# Install Required Python Packages

## Add to your Requirements File

In your main project section of Atom, you will create a file called `requirements.txt` and within it add the following lines:
- `django==2.2`
- `djangorestframework==3.9.2`

Originally as of 2018 he used 1.11 and 3.6.2 respectively. (Screenshots show these versions.)

## Install In Virtual Environment

### With Requirements File

In your **git bash** window, ensure your Vagrant file is up, connected, you are in the `/vagrant/` folder, and then run:

`pip install -r requirements.txt`

### Without the Requirements File

If you do not have the requirements file, you would do the following:

1. Have your **git bash** open, in the project directory, with vagrant server on and inside the virtual environment you intended to be in.

2. `pip install django==2.2` - it's important to install this specific version of django as a different version may have some differences.

<img src='files/IMGs/install/lang-01.png'>

3. `pip install djangorestframework==3.9.2`

<img src='files/IMGs/install/lang-02.png'>

#### Saving Our Requirements

It is standard practice for any python project to create this requirements file.

The `requirements.txt` contains a list of all of the python packages that your application uses & version number installed. Thus if you need to run on a different machine, you can use this file to reinstall all of the required packages and their exact version.

1. Be logged into the server via **git bash** program and in the correct virtual environment.

2. Switch to the vagrant folder:  `cd /vagrant/`

3. See requirements installed:  `pip freeze`

<img src='files/IMGs/django/setup03-requirements.png'>

The image above was supposedly default, but mine did not have some of these.

4. Pipe the output to a file:  `pip freeze > requirements.txt`

# Enable App In Django Settings File

In order to enable an app in Django, you will go to the root project folder - in this case, it is the **profiles_project** folder. From there, open the **settings.py** file - it is the configurations (or settings) file for the Django project.

Inside this section is the **INSTALLED_APPS** section where you will list all of the "apps" you need to use for your project. They can be installed via an external dependency (in the _requirements.txt_ file) or by creating them & adding to the installed apps list.

<hr>

In the settings file under **INSTALLED_APPS**, the strings are the apps relating to the project. Add (or enable) the following:

1. Add a new row (adding a string to a list) then add:  `rest_framework`

    This is required. It was installed earlier & is the name of the app within the dependency Django rest framework package necessary to add.
    

2. Add a new row (adding a string to a list) then add:  `rest_framework.authtoken`

    This is optional but should be included for security within your app. It will allow us to use the authentication token functionality that comes out of the box with the REST framework.


3. Add a new row (adding a string to a list) then add your manually created app:  `profiles_api`

**REMEMBER:** If you need to add any apps, you *must* add them here for them to be utilized by the project.

# Test And Commit Our Changes

We will test this using the **Django Development Server**. It is a very basic web server provided by django for the purpose of testing our work as we go along.

1. On **git bash** ensure the server is up, you're in the right project, and using the proper virtual environment.
<br>

2. Change directory to the **Src** folder:  `cd /vagrant/src/profiles_project`

If you run `ls` you will see several files, including the *manage.py* file. This file consists of several helper functions.

3. Run the following for all network interfaces:  `python manage.py runserver 0.0.0.0:8080`

This means don't listen on any particular IP address. So all IP addresses on the server will be able to access this application.

Run it on port 8080.
