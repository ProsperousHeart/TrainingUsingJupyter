If you previously created a baseline, you can review changes by going to_View_ ribbon & selecting **Gantt Chart > Tracking Gantt**. You may then see something like this, with the grey bars being the baseline:

<img src="IMGS/2/15.png">

You will want to reset the baseline before putting in actuals.

## Reset Baseline

1. Go to _Project_ ribbon and select **Set Baseline > Set Baseline**


2. To save original baseline, choose from a different dropdown in the **Set baseline** dropdown


## Update Gantt View

1. Go to _Format_ ribbon and select **Baseline**


2. Choose the newly saved baseline from the dropdown

## Update View So We Can See Status Moving Forward

1. Right-click in the Gantt chart area and select **Gridlines**


2. Scroll down to the bottom of the list and select **Status Date**
<img src="IMGS/2/16.png">


3. Ensure the **Type** is a solid line and choose a color that will stand out


4. Set status date by going to _Project_ ribbon, and at the **Status Date** section choose a date
<img src="IMGS/2/17.png">


5. A single line should appear on your Gantt chart
<img src="IMGS/2/18.png">


Goal of status date is to see what was in teh past and what is in the future.

## Entering Actuals

There are a few different ways to do this ...

### Task Ribbon - Percent Values

<img src="IMGS/2/19.png">

These will always be percent complete of duration.

1. Select your task(s)


2. Choose one of the percent values as shown above


This will show them complete in the information column with a bleu check, as well as captures teh actual start/finish in it's copy.


### Table For Entering Start/Finish

1. Go to _View_ ribbon, then choose **Tables** and select **Tracking**


2. Update as needed


You can also add in the **Start** and **Finish** that was planned by adding the columns. This way you can see the cmparison between current start/finish as well as actual.


## Task Usage

You can see better task detail by going to the **Task Usage** field.

1. Select your task


2. In the _View_ ribbon, select **Task Usage**


3. In the right hand table, right-click and seelect **Actual work**


4.  Within that same menu option for the right table, you can choose **Detail Styles**, add in **Overtime** field & check teh **Show In Menu** option


From here, you can view how work was delivered.
