# [Understanding Crashing](https://www.linkedin.com/learning/managing-time-constrained-projects-with-microsoft-project/understanding-crashing?autoSkip=true&autoplay=true&contextUrn=urn%3Ali%3AlyndaLearningPath%3A56db2e7e3dd5596be4e4989e&resume=false&u=76276956)

In this video, they used the **Exercise Files > Ch03 > Home_remodel_TC03_03.mpp** file.

The idea behind "crashing" a schedule is to see if there is additional time that can be saved based on resource requirements. In order to implement this, custom fields & tables will be needed.

## Create "Crash Table"

1. Go to _View_ ribbon and select **Tables > More Tables**


2. Click the **New...** button and call it `Crash Table`


3. Insert the following fields:
    - `ID`
    - `Name` with width `20`
    - `Duration`
    - `Cost`


4. Create additional fields to capture new data (rename them later):
    - `Duration1`
    - `Cost1`
    - `Number1`
    - `Cost2`
    - `Cost3`
    - `Fixed Cost`


## Create View That Utilizes Table

1. In _View_ ribbon, click **Other Views > More Views**


2. Ensure __Gantt chart__ is selected and hit the **Copy...** button - rename this new view as `Crash View`


3. Make the following changes to the view:
    - Table:  `Crash Table`
    - Filter:  `Critical`


4. Hit the **OK** then **Apply** buttons

## Rename Fields / Columns

Right-click on the column (any column) and choose **Custom Fields**, select the field, and hit the **Rename...** button. Renamethe following:

- `Duration1` to `Crash duration`
- `Cost1` to `Crash Cost`
- `Number1` to `Crash Dur Delta` (a formula - difference between current and crash duration)
- `Cost2` to `Crash Cost Delta`
- `Cost3` to `Crash Cost per day`

**NOTE:**  You can move between field types by changing the **Type** in the top-right corner.

## Build Formulas For Fields

While still in the **Custom Fields** box ...

### Update "Crash Dur Delta"

1. Be sure you are in the **Number** field section -0 choose the **Crash Dur Delta** field


2. In **Custom Attributes** choose **Formula** button
<img src="IMGS/2/08.png">


3. After hitting the **Function** button choose **General > Expression > IIif**
<img src="IMGS/2/09.png">


4. Update the formula (within the parentheses) to be:  `[Duration1]>0,([Duration]-[Duration1])/480,"0"`
    The 480 turns it into days.


5. Once you hit the **OK** button, you will see the following:
<img src="IMGS/2/10.png">


### Update "Crash Dur Delta"

1. Be sure you are in the **Cost** field section - choose the **Crash Cost Delta** field


2. In **Custom Attributes** choose **Formula** button


3. After hitting the **Function** button choose **General > Expression > IIif**


4. Update the formula (within the parentheses) to be:  `[Duration1]>0,[Cost1]-[Cost],"0"`


You will be prompted by the bulletin like before once you hit the **OK** button.


### Update "Crash Per Day"

1. Be sure you are in the **Cost** field section - choose the **Crash Cost per day** field


2. In **Custom Attributes** choose **Formula** button


3. After hitting the **Function** button choose **General > Expression > IIif**


4. Update the formula (within the parentheses) to be:  `[Number1]>0,[Cost2]/[Number1],"0"`


You will be prompted by the bulletin like before once you hit the **OK** button.
