# What Is pandas?

[Pandas](https://www.python-course.eu/pandas.php) is an acronymn & stands for "Panel Data Structures". This is used by economists, statisticians, econometricians, etc.

Pandas is a library built **on top** of NumPy, which is built on the [Python programming language](https://github.com/ProsperousHeart/TrainingUsingJupyter/tree/master/Python).

Libraries are a bundle of features and add-ons to the base language. Each one has it's own focus (specialty). 

For **pandas**? It's <font color=green>data analysis</font>.

It is a robust toolkit analyzing, filtering, manipulating, aggregating, merging, pivoting, and cleaning data.

It's also been known as *"Excel for Python"* or even "Excel on steroids".

The reason why it is better than Excel or Google Sheets is because it can calculate data on 10s of 1,000s of rows of data within **seconds**.

Additional items of note:
- Pandas is an open source library built on top of NumPy
- It allows for fast analysis & data cleaning/preparation
- exxcels in performance & productivity- built-in visualization features
- can work with data from a wide variety of sources

## Installation Basics

The pandas library is dependent on other python libraries in order to run, so it's just easiest to install the "suite".

It is **_highly recommended_** you install Python using the [Anaconda distribution](https://anaconda.org) to make sure all underlying dependencies (such as Linear Algebra libraries) all sync up with the use of a conda install. It bundles:

If installing to your terminal ([Anaconda](#anaconda-installation) should already have them) you can install pandas by going to your terminal or command prompt and typing one of the following (depending on your environment):
    
    conda install pandas

If you do not have Anaconda and can not install it, **please refer to [Numpy's official documentation on various installation instructions.](http://docs.scipy.org/doc/numpy-1.10.1/user/install.html)**

## Activating Root Environment

When Anaconda is installed, it creates a default environment called **root** (like a sandbox). This is the starting point for all other sandboxes.

The reason it supports multiple "sandboxes" is so that different users can play around with different versions of libraries and languages. It also allows the user to see how different libraries play with each other.

To activate this sandbox in CMD, input:  `activate root`

Otherwise, if you are in the Anaconda prompt you are already in the base environment.

<img src="IMGs/ACMD-01.png">

## Using Conda Library In Anaconda

This is one of the 100s of libraries within Anaconda. It is used for updating & installing other python libraries.

### Update Libraries In Conda

In your prompt, put in:  `conda update conda`

This will query the internet to see if any updates are available. If some are found, it will provide a message similar to this:

<img src='IMGs/ACMD-02.png'>

Be sure to press **y** so it is updated.

If you get a permissions error, try running the Anaconda prompt as administrator.

You can do this with any library:  `conda update pandas`

There may be times other libraries need to be updated - at that point they will appear above the one you requested to update.

If it's already up to date, it will let you know it is already installed.

<img src='IMGs/ACMD-03.png'>

What if you wanted to update all libraries? Simnple.
`conda update --all`

At the time I went through this, there were a LOT of packages that needed to be updated.

When you run the command to update all, it will tell you:
- which packages will be downloaded (showing package, build, and size)
- new packages to be installed
- which packages will be updated

The instructor suggested to do this at minimum once a month, if not every week. This will ensure you are on the latest and greatest versions.

### Install Libraries In Conda

If the library is already installed, there will be no errors. Simple a message informing you that the package is already installed.

Suggested items to install:
- `conda install markdown`
- `conda install numexpr` (used for optimization & speed efficiencies in pandas behind the scenes)
- `conda install bottleneck` (used for optimization & speed efficiencies in pandas behind the scenes)
- `conda install pandas-datareader` (used to fetch financial data from the internet)
- `conda install xlsxwriter` (used for opening and writing to XLSX files)
- `conda install openpyxl`
