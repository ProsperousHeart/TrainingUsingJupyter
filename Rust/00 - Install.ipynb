# Installing Rust

## Mac

1. `brew install rustup init`
2. `brew install --cask visual-studio-code`

## Linux

`curl --proto '=https' --tlsvl.2 -sSf https://sh.rustup.rs | sh`





.. OR use package manager, then install:  `rustup toolchain install stable`

## Windows

1. **rustup** - manages Rust installation (will download & manage tool chain to work on your system)

2. **Vistual Studio Vode** - coding environment

3. **MSVC C++ Built Tools** - needed to build on Windows

### VS Code

These instructions were given to me through a training, but I could not install on Windows without VS Code. So this may not be applicable for you.

So if you don't already have it ...

#### Installing The Software

1. Download from [code.visualstudio.com](https://code.visualstudio.com/) - default settings are fine

2. You'll need to log out or restart your machine


### Running First Program

1. Open VS Code and go to **Terminal > Net TErminal**

2. Run `cargo init hello`



3. Go to **File > Open Folder > hello**





4. Editor will reload.



5. Go to **Terminal > New Terminal** and send `cargo run`



Program should compile and run - if Rust is installed.
